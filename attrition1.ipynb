## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition", "Department"]]
y_df
```

```python
# Create a list of at least 10 column names to use as X data
X_data = attrition_df[['Education', 'Age', 'DistanceFromHome', 'JobSatisfaction', 'OverTime', 
                     'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 
                     'NumCompaniesWorked']]


# Create X_df using your selected columns
X_df = pd.DataFrame(X_data)

# Show the data types for X_df
X_df.dtypes
```

```python
X_df
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

#Convert 'Yes' to 1 and 'No' to 0
X_df['OverTime'] = X_df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)

#Convert the column to int64
X_df['OverTime'] = X_df['OverTime'].astype('int64')
```

```python
# Convert 'Yes' to 1 and 'No' to 0
y_df['Attrition'] = y_df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Convert the column to int64
y_df['Attrition'] = y_df['Attrition'].astype('int64')
```

```python
print(y_df)
print(X_df)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=19)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
ohe.fit(y_train[['Department']])
feature_names = ohe.get_feature_names_out()

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded = pd.DataFrame(ohe.fit_transform(y_train[['Department']]), columns=feature_names)
y_test_encoded = pd.DataFrame(ohe.transform(y_test[['Department']]), columns=feature_names)

```

```python
from sklearn.preprocessing import LabelEncoder

# Create a OneHotEncoder for the Attrition column
le = LabelEncoder()

# Fit the encoder to the training data
le.fit(y_train['Attrition'])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded2 = pd.DataFrame(le.transform(y_train[['Attrition']]))
y_test_encoded2 = pd.DataFrame(le.transform(y_test[['Attrition']]))
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_features = X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(num_features,), name='Input_Layer')

# Create at least two shared layers
shared_layer1 = layers.Dense(128, activation='relu', name='Share_Layer1')(input_layer)
shared_layer2 = layers.Dense(256, activation='relu', name='Shared_Layer2')(shared_layer1)
shared_layer3 = layers.Dense(256, activation='relu', name='Shared_Layer3')(shared_layer2)
shared_layer4 = layers.Dense(128, activation='relu', name='Shared_Layer4')(shared_layer3)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden1 = layers.Dense(128, activation='relu', name='Dept_Hidden1')(shared_layer4)
dept_hidden2 = layers.Dense(64, activation='relu', name='Dept_Hidden2')(dept_hidden1)

# Create the output layer
dept_output = layers.Dense(3, activation='softmax', name='Department')(dept_hidden2)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden1 = layers.Dense(128, activation='relu', name='Attrition_Hidden1')(shared_layer4)
attrition_hidden2 = layers.Dense(64, activation='relu', name='Attrition_Hidden2')(attrition_hidden1)

# Create the output layer
attrition_output = layers.Dense(1, activation='sigmoid', name='Attrition')(attrition_hidden2)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', 
             loss={ "Department": 'categorical_crossentropy', 'Attrition': 'binary_crossentropy'},
             metrics={'Department': 'accuracy', 'Attrition': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {'Department': y_train_encoded, 'Attrition': y_train_encoded2},
    epochs=2500,
    batch_size=32,
    validation_split=0.2)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'Department': y_test_encoded, 'Attrition': y_test_encoded2})

# Extract the test loss and test accuracy from the results
loss = test_results[0]
(f'Loss: {loss:.2f}')
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {test_results[1]:.2f}")
print(f"Attrition Accuracy: {test_results[2]:.2f}")
```
